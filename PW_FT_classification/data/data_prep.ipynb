```python
# Paths to your CSV files
csv_paths = [
    "/media/mo/nvme0n1/MD_up_TT/TT csv files/train_bb_20250204_094704.csv",
    "/media/mo/nvme0n1/MD_up_TT/TT csv files/val_bb_20250204_094704.csv",
    "/media/mo/nvme0n1/MD_up_TT/TT csv files/test_bb_20250204_094704.csv"
]
# 1. Read & combine all CSVs
dfs = [pd.read_csv(path, low_memory=False) for path in csv_paths]
# train, val and test concatenated
df_tvt = pd.concat(dfs, ignore_index=True)
df_tvt.shape
```

```python
import re
df_tvt_like_temp = df_tvt.copy()

# add these columns to the df_tvt: path', 'classification', 'label', 'Photo_Time', 'Location'

# create a 'path' column same as the 'image' column
df_tvt_like_temp['path'] = df_tvt['image']

# use the 'image_label_1' column for 'label'
df_tvt_like_temp['label'] = df_tvt['image_label_1']

# create a dummy numerical code for the label column (using categorical codes)
df_tvt_like_temp['classification'] = df_tvt['image_label_1'].astype('category').cat.codes

# assign the 'timestamp' column to 'Photo_Time'
df_tvt_like_temp['Photo_Time'] = df_tvt['timestamp']

# extract location from the 'image' path; e.g., extract "HG/O15/CAM136" from "HG/O15/CAM136/20230630/101_BTCF/IMG_0244.JPG"
def robust_extract_location(path):
    tokens = path.split('/')
    # Find the token that starts with "CAM"
    for i, token in enumerate(tokens):
        if token.startswith('CAM'):
            # Return all tokens from the start up to and including the camera token
            return '/'.join(tokens[:i+1])
    return None

df_tvt_like_temp['Location'] = df_tvt['image'].apply(robust_extract_location)
df_tvt_like_temp.columns
```
